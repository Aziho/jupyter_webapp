```python
import pandas as pd
bikes=pd.read_csv('bikes.csv')
```

```python
bikes.describe()
```

```python
bikes.plot(kind='scatter', x='temperature', y='rentals')
```

```python
bikes.plot(kind='scatter', x='humidity', y='rentals')
```

```python
response='rentals'
Y=bikes[[response]]
Y
```

```python
predictors=list(bikes.columns)
predictors.remove(response)
X=bikes[predictors]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(X, Y, random_state=1234)
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression().fit(x_train,y_train)
model
```

```python
#model_intercept
model.intercept_
```

```python
#model slop
model.coef_
# The slopes corespond to the order of the independent variables which are listed in the training data
```

```python
#Evaluate the model
model.score(x_test,y_test)
```

```python
y_predict= model.predict(x_test)
from sklearn.metrics import mean_absolute_error
mean_absolute_error(y_test, y_predict)
```
